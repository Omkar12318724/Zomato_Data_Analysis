# **Zomato Data Analysis Project**

**Step -1 Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
dataframe = pd.read_csv('/content/Zomato data .csv')
print(dataframe)
```

```python
dataframe
```

### **convert the data type of colume - rate**

```python
def handleRate(value):
  value=str(value).split('/')
  value=value[0];
  return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

**Type Of Resturant**

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel('Type Of Resturant')
plt.ylabel('Count')
```

**conclusion majority of the resturent falls in dinning catagory**

```python
dataframe.head()
```

```python
grouped_data=dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result, c="green", marker = "o")
plt.xlabel("Type Of Resturant", c="red", size =20)
plt.ylabel("votes", c="red", size=20)
```

**conclusion- Dinning Resturant recived maximum Votes**

```python
plt.hist(dataframe['rate'], bins =40)
plt.title('Distribution of Ratings')
plt.show()
```

**majority Resturant Recived rating from 3.5 to 4**

```python
dataframe.head()
```

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

**Conclusion -Majority of couples prefeeres resturant of cost 300 rupees**

```python
plt.figure(figsize=(6,6))
sns.boxplot(x = 'online_order', y ='rate', data =dataframe)
plt.show()
```

**Conclusion - offline order reciverd lower rating comapre to online**

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')
plt.title("heatmap")
plt.xlabel("online_order")
plt.ylabel("listed_in(type)")
plt.show()
```
